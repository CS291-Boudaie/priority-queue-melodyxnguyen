# Notebook State
Unlike a Python program, which will run from the first line to the last line, a Jupyter notebook is executed based on the order of cells that you run. This is all managed in the __kernel__; the process for this notebook. For example, try the following:

Run Cell A (notice how you get an error), then run Cell B and then Cell A again.

Keep running Cell A and see how the value of `a` keeps going up. Then run Cell B again and Cell A after. Notice how the vale of `a` is reset. Make sure you understand why.

```python
# CELL A
a += 1
print(a)
```

```python
# CELL B
a = 0
print(a)
```

There are a few tools you should be aware of to manage notebook state. If you go to the top toolbar (you will see File/Edit/View...), here are some commonly used tools:

1. Kernel/Interrupt Kernel: If the notebook gets stuck for whatever reason (e.g. infinite loop), you can run this to stop whatever the kernel is doing.
2. Restart Kernel and Clear Outputs of all cells: Resets the kernel back to a blank slate, and removes the outputs from the cells you ran. Good for cleaning up after a lot of experimentation.
3. Restart Kernel and Run All Cells: Does exactly what it says; good for testing your notebook from the start to the end

# Jupyter "Magic"

There are a few "magic" commands that make programming in the environment easier. You can see a full list of commands here: https://ipython.readthedocs.io/en/stable/interactive/magics.html

One favorite is `%%time` which times how long a cell takes to run, or `%time` which times how long a line takes to run.

```python
# Timing Commands
# We always hear that hash sets are faster than arrays for lookups, let's test if that's true
import random
def lookup_set(num_additions, num_lookups):
    s = set()
    # Add 100k items to the list
    num_hits = 0
    for i in range(num_additions):
        s.add(random.randint(1,10_000_000))

    for i in range(num_lookups):
        if random.randint(1,10_000_000) in s:
            num_hits += 1
    return num_hits

def lookup_list(num_additions, num_lookups):
    s = []
    # Add 100k items to the list
    num_hits = 0
    for i in range(num_additions):
        s.append(random.randint(1,10_000_000))

    for i in range(num_lookups):
        if random.randint(1,10_000_000) in s:
            num_hits += 1
    return num_hits
```

```python
%%time 
lookup_set(100_000, 10_000)
```

```python
%%time 
lookup_list(100_000, 10_000)
```
